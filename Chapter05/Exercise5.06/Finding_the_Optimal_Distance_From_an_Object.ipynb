```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
x = np.linspace(0.1, 50, 2000)
y = np.arctan(7*x / (30+x**2) )
plt.plot(x,y)

plt.show()
```

```python
ymax = max(y)
xmax = x[list(y).index(ymax)]

print(round(xmax,2), round(ymax,2))
```

```python
ymax_degrees = round(ymax * 180 / np.pi, 2)
print(ymax_degrees)
```
