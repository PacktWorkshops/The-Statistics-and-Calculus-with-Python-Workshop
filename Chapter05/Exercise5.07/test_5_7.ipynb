```python
import unittest
import import_ipynb
import numpy as np
from Visualizing_Vectors import *

class Test(unittest.TestCase):
    
    def test_find_angle(self):
        vec1 = np.array([1,5])
        vec2 = np.array([5,-1])
        self.assertEqual(find_angle(vec1, vec2, isRadians = False, isSum = True), 90.0)
 
        vec1 = [10,2]
        vec2 = [-2.8,-1]
        self.assertEqual(find_angle(vec1, vec2, isRadians = True, isSum = False), 2.9959642730210043)

if __name__ == '__main__':
    unittest.main(argv=['first-arg-is-ignored'], exit=False)
```
