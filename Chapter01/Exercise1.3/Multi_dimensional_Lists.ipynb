```python
a = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
```

```python
for row in a:
    print(row[0])
```

```python
for row in a:
    for element in row:
        print(element)
```

```python
for i in range(3):
    print(a[i][i])
```

```python
for i in range(3):
    print(f'The {i + 1}-th diagonal element is: {a[i][i]}')
```
